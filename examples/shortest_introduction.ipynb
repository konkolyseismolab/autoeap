## Example: create simple raw light curve using your own Target Pixel File (TPF)

TPF are available in FITS format in the Kepler Archive or MAST database.

As an example, we provide the C17 TPF of EPIC 212466080, here: https://github.com/zabop/autoeap/tree/master/examples

```python
from autoeap import createlightcurve

inputfile = 'ktwo212466080-c17_lpd-targ.fits'
```

```python
time,flux,flux_err = createlightcurve(
        inputfile,             # Path to TPF or EPIC ID
        apply_K2SC=False,      # Apply K2 systematic correction
        remove_spline=False,   # Fit and remove low-order spline
        save_lc=False,         # Save final light curve to file
        campaign=None,         # K2 campaign number if TPF should be downladed
        show_plots=False,      # Vizualize steps inside notebook
        save_plots=False,      # Save plot files in a folder
        polyorder='auto',      # The order of the detrending polynomial
        sigma_detrend=10       # The number of standard deviations to use for sigma clipping limit before spline correction.
        )
```

#### Visualize the result

```python
import matplotlib.pyplot as plt

plt.figure(dpi=100)
plt.title('Raw photometry')
plt.scatter(time, flux, s=1, color='black')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.show()
plt.close()
```

## Example: download TPF automatically

If you want to let autoEAP download TPF from MAST, you can specify the target ID and the obervation campaign (if there are more than one).

```python
from autoeap import createlightcurve

targetID = 'EPIC 212466080'
campaign = 17
```

```python
time,flux,flux_err = createlightcurve(
        targetID,              # Path to TPF or EPIC ID
        apply_K2SC=False,      # Apply K2 systematic correction
        remove_spline=False,   # Fit and remove low-order spline
        save_lc=False,         # Save final light curve to file
        campaign=campaign,     # K2 campaign number if TPF should be downladed
        show_plots=False,      # Vizualize steps inside notebook
        save_plots=False,      # Save plot files in a folder
        polyorder='auto',      # The order of the detrending polynomial
        sigma_detrend=10       # The number of standard deviations to use for sigma clipping limit before spline correction.
        )
```

#### Visualize the result

```python
import matplotlib.pyplot as plt

plt.figure(dpi=100)
plt.title('Raw photometry')
plt.scatter(time, flux, s=1, color='black')
plt.xlabel('Time')
plt.ylabel('Flux')
plt.show()
plt.close()
```
